# Create a plot with blended lineproperties

```python
from xarrayutils.plotting import fancy_plot
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Lets produce a test dataset

```python
x = np.linspace(0, 4*np.pi, 50)
y = np.cos(x)
```

Now we can plot the line...

```python
plt.plot(x,y)
```

But what if we want to adjust one of the line parameters, like `linewidth` or `alpha` *along* the line?

This works nicely with markersize and linewidth

```python
prop = np.linspace(0,1, len(x))
prop2 = -np.linspace(-1,0, len(x))
fancy_plot(x,y,{'linewidth':(prop)*4, 'markersize':prop*prop2*60,}, marker='.')
```

And for colors...

```python
color_gray = [str(a) for a in prop]
fancy_plot(x,y,{'color':color_gray})
```

But when varying the transparency, the overlapping segments ruin the effect.

```python
fancy_plot(x,y,{'alpha':prop}, linewidth=4)
```

An intermediate solution is to interpolate the linesegments, so that there is a lot of overlap, effectively creating a gradient.

```python
x = np.linspace(0, 4*np.pi, 500)
y = np.cos(x)
prop = np.linspace(0,1, len(x))
```

```python
fancy_plot(x,y,{'alpha':prop}, linewidth=4)
```

But this makes the whole plot less transparent and can also increase the computational burden.
